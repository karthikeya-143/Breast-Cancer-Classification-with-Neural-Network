```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
 print(breast_cancer_dataset)
```

```python
df=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
df["label"]=breast_cancer_dataset.target
```

```python
X=df.drop(columns="label",axis=1)
Y=df["label"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Building the NN
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of Neural Network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(20,activation="relu"),
    keras.layers.Dense(2,activation="sigmoid")
])
```

```python
#Compiling the Neural Network
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
#Training the Neural Network
history=model.fit(X_train,Y_train,validation_split=0.1,epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["train","validation"],loc="upper left")
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["train","validation"],loc="upper left")
```

Accuracy of the model on test data

```python
loss,accuracy=model.evaluate(X_test,Y_test)
print(accuracy)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Example: assume X_train is your training feature matrix
scaler = StandardScaler()
scaler.fit(X_train)   # <-- Fit on training data

# Now test with one input row (30 features from breast cancer dataset)
input_data = (17.99, 10.38, 122.8, 1001.0, 0.1184, 0.2776, 0.3001, 0.1471,
              0.2419, 0.07871, 1.095, 0.9053, 8.589, 153.4, 0.006399,
              0.04904, 0.05373, 0.01587, 0.03003, 0.006193, 25.38, 17.33,
              184.6, 2019.0, 0.1622, 0.6656, 0.7119, 0.2654, 0.4601, 0.1189)

# Convert to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape for single instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize using the fitted scaler
input_data_std = scaler.transform(input_data_reshaped)

# Predict
prediction = model.predict(input_data_std)
print(prediction)

prediction_labels = [np.argmax(prediction)]
print(prediction_labels)

if prediction_labels[0] == 0:
    print("The breast cancer is Malignant")
else:
    print("The breast cancer is Benign")
```
